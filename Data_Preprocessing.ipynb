Extracting Valid Characters from fund names using regexes
Decreasing size of data to avoid crashing Protege


```python
import pandas as pd
import re

# Load your CSV file
df = pd.read_csv("/content/drive/MyDrive/Data_NLP/fossilfree.csv")  # Replace with the actual file path

# Define a function to clean FundName
def clean_fund_name(name):
    if pd.isna(name):  # Check for NA
        return name
    # Replace spaces with underscores
    name = name.replace(" ", "_")
    # Remove any non-alphanumeric characters, underscores, hyphens, and slashes
    name = re.sub(r'[^A-Za-z0-9_/.-]', '', name)
    return name

# Apply the cleaning function to the FundName column
df['FundName'] = df['FundName'].apply(clean_fund_name)

# Drop rows with any missing values
df = df.dropna()

# Convert the 'Weight' column to numeric
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')

# Save the cleaned dataset to a new CSV file
df.to_csv("/content/drive/MyDrive/Data_NLP/fossilfree.csv", index=False)  # Saves the cleaned CSV to a new file
```
